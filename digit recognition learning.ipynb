### Lakukan dengan menggunakan miu=0.1, layer =3, 5, 10, 15, 20, 25

```python
import math
import matplotlib.pyplot as plt
from importlib import reload

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier as MLP

import warnings
warnings.filterwarnings('ignore')

dataset = pd.read_csv("E:/PENS/Semester5/Neuro Computing/Praktikum/Data/digit.csv")
data = dataset.iloc[:, 0:-1]
label = dataset.iloc[:, -1]

# number_to_display = 80
# num_cells = math.ceil(math.sqrt(number_to_display))
# plt.figure(figsize=(20,20))

# for i in range(number_to_display):
#     ax = plt.subplot(num_cells,num_cells , i+1)
#     ax.imshow(data.iloc[i].values.reshape(10,10))
#     #plt.xlabel=''
#     ax.axes.get_xaxis().set_visible(False)
#     ax.axes.get_yaxis().set_visible(False)
#     plt.title(label.iloc[i])

# plt.show()

layer = np.array([3, 5, 10, 15, 20, 25])

def mse(actual, pred):
    actual, pred = np.array(actual), np.array(pred)
    return np.square(np.subtract(actual, pred)).mean()

for i in range(0,6):
    model = MLP(tol=0, hidden_layer_sizes=layer[i], activation='logistic', 
                learning_rate_init=0.1)
    model.fit(data,label)
    pred = model.predict(data)
    if i == 0:
        error = np.array(mse(label, pred))
    else:
        error = np.append(error, mse(label, pred))
    
    print('Using hidden layer:',layer[i])
    print('Score:', model.score(data, label))
    
    plt.plot(model.loss_curve_)
    plt.xlabel('Iterations')
    plt.ylabel('Loss')
    plt.title('Gradient Descent Digit')
    plt.show()
    
    model.fit(data, label)
    test_accuracy = model.score(data, label)
    print(f"Akurasi testing data : {test_accuracy * 100:.2f}%")

    predictions = model.predict(data)
    print(f"Prediksi data test: {predictions}\n\n")
        
    
```
